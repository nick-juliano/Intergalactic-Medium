# Warm-Hot Intergalactic Medium
___
## Introduction

By the 1990s, the <b>missing baryon problem</b> was an unexplained difference between the current mean baryon density derived from the results of Big Bang Nucleosynthesis and the density of detected baryons in galaxies, clusters, and Lyman alpha systems by about a factor of 2. At the time, it was generally believed that the unobserved baryons were in a hot, low-density gas spread, but it was nontheless difficult to detect.

This medium of intergalactic space is refered to as the warm-hot intergalactic medium (WHIM). The WHIM is the filamentary structure, or "cosmic web" observed in numerical cosmological simulations.

## 10.1 Simulations

- The figure below is from "Illustris TNG100",  a numerical cosmological simulation that that provides insight into the proporties of the warm-hot intergalactic medium. 
- Cubic box with sides of comoving length $l \sim 110  Mpc$, a $\Lambda$CDM universe with <i>Planck</i> 2016 results.
- The two boxes in the figure are from $z=0$, or present day. 
    - Left panel: distribution of diffuse IGM
        - Widely distributed
    - Right panel: distribution of WHIM
        - Found primarily in long filaments. 
- The figure shows that the WHIM is shocked and heated upon the impact of meeting filaments. As the universe develops with time, the amount of WHIM increases.
    - From recombination ($z=7$) to present day, the percentage of baryonic mass in the WHIM goes from 0.6% to 45%.
- Despite having a much lower density than hot ISM, WHIM maintains its high temperature due to a high cooling time. 


## 10.2 Observations

- Free-free emission dominates, and observing free-free emission from WHIM is difficult because neutral hydrogen in the Milky Way prevents most extragalactic free-free emission from reaching us.
- The neutral hydrogen in our galaxy differs in optical depth by ~2 orders of magnitude, so there are better and worse lines of sight for observing WHIM.
- Alternatively, we can look for absorption and emission lines from highly ionized metals. The O VI doublet can be seen in absorption toward hot white dwarfs within the Local Bubble of hot ionized gas. Thus, a region of warm-hot ionized medium with a comparable temperature should produce O VI absorption lines in the spectra of background UV sources.
