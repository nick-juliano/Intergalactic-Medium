## Introduction

Inside virial theorem limits, the following mass estimate can be made

$M \approx \frac{\sigma^2 r}{G}$

$\sigma \equiv$ velocity dispersion of material inside radius $r$

$r$ is commonly approximated as $r=r_{200}$ within which $\bar{\rho}_{200}=200\rho_{c,0}$, where $\rho_{c,0}\equiv\frac{3H_{0}^{2}}{8\pi G}=9.20 \times 10^{-30} g\ cm^{-3}h_{70}^{2}$, the critical density at which the universe is flat.

$r_{200,MW}\approx 220kpc$. For context, the scale length of the Galaxy's stellar disk is $h_R \approx 3 kpc$
___

## 8.1 Circumgalacitc Medium: Our Galaxy

- $r_{200} \approx 220 kpc$ is approximately 1/3 the distance to the Andromeda Galaxy.
- large variety of gaseous components.
- Gas falling in from the intergalactic medium, stripped from satellite galaxies, ejected from the interstellar medium of central portion of galaxy.
- First detected 21cm emission. Discrete HI clouds with RVs higher than expected for typical 21cm ISM. (Origin of name <i>high-velocity cloud</i>) SEE FIGURE 8.1
- Magellanic Stream - high-velocity gas that has been stripped from the Magellanic Clouds by tidal effects and ram-pressure stripping.
    - galactocentric velocity $v_{gal}$ ranges from +90 $km\ s^{-1}$ to -240 $km\ s^{-1}$
    - $M_{HI}\sim 3 \times 10^{8}M_\odot$
    - $M_{ms,tot}\sim 1\times10^{9}M_\odot$
    - $M_{ms,tot}$ is similar to remaining gas in Magellanic Clouds
- Other HVCs
    - Not clearly associated with other satellite galaxies.
    - galactocentric velocity $v_{gal}$ ranges from +250 $km\ s^{-1}$ to -250 $km\ s^{-1}$
    - $T_{gas} \sim 9000 K$
    - $\sim 6%$ of our sky is covered by HVCs with $N_{HI}>10^{19}cm^{-2}$
    - $\sim 0.3%$ is covered by HVCs with $N_{HI}>10^{20}cm^{-20}$ 
- Distances 
    - Determined by "sandwiching" HVCs between halo stars of known distances to constrain bounds. 
    - Most $d<15kpc$
    - Within $\approx 30 ^{\circ}$ of the disk plane
    - $\Rightarrow$ associated with disk of galaxy rather than throughout dark halo.
    **Question: "With distances known, we can determine physical sizes and masses of individual HVCs" p184. If we just have constraints on distances, isn't this statement based on an unjustified assumption?**
- Masses
    - Using distances, we can determine masses
    - HI mass of HVCs, $M_{HI} \sim 3 \times 10^{7}M_{\odot}$
    - With helium and ionized hydrogen, $M_{hvc} \sim 7 \times 10^{7} M_{\odot}$, noting that $M_{hvc}<0.1M_{ms}$

## 8.2 Circumgalacitc Medium: Other Galaxies
Two sources of CGM gas:

1) Low-metallicity intergalactic gas inflow
    - Shown via numerical simulations of structure formation in $\Lambda CDM$ cosmology
    - Filamentary flow, dilutes metallicty of CGM
2) Galactic wind outflow from interstellar gas.
    - Varying degrees of galactic wind strength
    - Powered by supernova explosions (in the case of M82)

## 8.3 Intracluster Medium

$L_{MW, V} \sim 2 \times 10^{10}L_{\odot, V}$, 

$L_{M31, V} \sim 1.3 L_{MW, V}$,

$L_{LG, V} \sim 2.7 L_{MW, V}$

$\Rightarrow L_{MW, V} + L_{M31, V} \sim 0.85 L_{LG, V}$

Local Group contains >50 galaxies, most with $L_{V} < 10^{-4} L_{MW, V}$

The nearest rich cluster is the Coma Cluster, at a distance of approximately 100 Mpc

## Key Definitions

<b> Circumgalactic Medium (CGM)</b> - The diffuse gas, plus any stray dust, that lies outside the main body of a galaxy's stellar distribution but inside the virial radius of the galaxy's dark halo.

<b> Intracluster Medium</b> - The diffuse hot gas that lies inside the virial radius of a cluster of galaxies but is not bound to any particular galaxy in the cluster.

<b> Virial Radius</b> - The radius of the largest sphere centered on the system's barycenter inside whih the steady-state virial theorem holds true. 

<b> Local Group</b> - The cluster of galaxies in which we live. Dominated by the Milky Way Galaxy and the Andromeda Galaxy.
