## Introduction

Much of the hydrogen et. al. has not accumulated in galaxies and remains in intergalactic space.

Temperature fluxtuations provide insight on the density of baryonic matter in the universe. <i>Planck</i> puts this value at $\bar{\rho}_{bary,0} = (4.21 \pm 0.03) \times 10^{-31} g\ cm^{-3}$, which places the mean number density of baryons at $\bar{n}_{bary,0} = 2.52 \times 10^{-7} cm^{-3}$
___

## 9.1 Gunn-Peterson Effect

- Because of universal expansion, certain parameters change with time/distance: 
    - Mean baryon density & number density decreases with time. $\bar{n}_{bary}(t)=\bar{n}_{bary,0}(1+z)^{3}$
    - Energy of emitted photons $hv(t)=hv_{0}(1+z)$
- Optical depth at redshift $\tau_{v}=c\sigma_\alpha \frac{n_{l}(z)}{H(z)}$
- At redshifts $z<<1$, $\tau_{v}\sim 1.5\times 10^4(\frac{n_{l,0}(z)}{\bar{n}_{bary,0}})$
- For a quasar at redshift $z$, strong Lyman $\alpha$ absorption in the observed wavelength range $\lambda_{0} 1216 \rightarrow 1216(1+z)  \unicode[serif]{xC5}$ is called the <b>Gunn-Peterson effect</b>
- Consequence of this effect, intergalactic gas either:
    - ⬜️ has a density much lower than mean baryon density of the universe
    - ✅ is very highly ionized

## 9.2 Recombination

- Hydrogen stopped beion ionized at recombination, $z \sim 1400$
- Hydrogen became ionized again at reionization, $z \sim 7$

- Temperature of CMB decreases with expansion. $kT_0 = (2.3486 \pm 0.0005 \times 10^{-4}eV$
- Temperature previously was $T=T_{0}(1+z)$
- When the temperature dropped below $kT \sim 66 keV$, deuterium nuclei could form without photodissociation. Enter Big Bang Nucleosynthesis. 
- At $kT=0.330eV$ (i.e., $z\sim1400$ and $t \sim 0.23 Myr$, recombination of hydrogen.
___

## 9.3 Reionization

- "Dark Ages" is brought by recombination of hydrogen: era between recombination and formation of first stars ($t\sim100Myr$).
- Not completely dark, CMB still exists. 
- <b>Reionization</b> accomplished by UV photons from massive stars and AGN.
- Patchy, gradual process around individual star-forming galaxies, expanding outward. So there is no clear instant of reionization and it is understood better as an epoch of transition.
- What the CMB can tell us about reionization:
    - low-redshift reionized gas of IGM informs the population of free elections between us and CMB.
    - free electrons scatter photons of CMB via Thomson scattering. 
    - If optical depth from Thomson scattering was high, then CMB temperature fluxuations would disipate and make original direction of motion untracable. 
    - Thus, $\tau _ {e} <<1$
    - The small amount of scatterings have imposed a slight feathering of the temperature gradient in the CMB.

## 9.4 Lyman Alpha Forest

This section interprets observed regions with varying column densities, defined in order.

- <b>Lyman alpha forest</b> - a portion of short-wavelength absorption lines observed along the line of sight to quasars (see figure 9.1)
    - Not representative of gravitationally collapsed systems. 
- <b>Lyman limit systems</b> - slightly more dense than a Lyman alpha forest, these systems are thick enough that radiation doesn't reach their interior and neutral hydrogen still remains (self shielding). 
- <b>Damped Lyman alpha systems</b> - absorbing systems with a large neutral hydrogen column density. 
    - These column densities are comparable to large galaxies.
    - Thought of as gravitationaly bound protogalaxies which are not effective at converting gas into stars. 

- <b>fluctuating Gunn-Peterson approximation </b> - the below equation (9.65) that describes the equivalent of Gunn-Peterson absorption for a non-uniform photoionized medium. 

$\tau = A(\frac{n}{\bar{n}_{bary}})^{1.58}$
